```python
import requests
import json

REQUEST = requests.post

HEADERS = {u'content-type': u'application/json'}

URL = "http://localhost:3000/createComment"

#Get token
token = requests.post("http://localhost:3000/isUser", headers=HEADERS, data=json.dumps({"username" : "tester", "passwordHash": "my dad"})).json()["token"]

def test(body):
    res = REQUEST(URL, headers= HEADERS, data=json.dumps(body), timeout=2.5)
    print("Code: {}".format(res.status_code))
    print("Msgs: {}\n".format(res.text))
    return res
```

```python
#Works
test({"username":"tester", "token":token, "text":"Hello database", "file":"tester-636120a629e1539d87fa47afee8847a253690437-1614799468655.pdf"})

test({"username":"tester", "token":token, "text":"Hello database! (again)", "file":"tester-636120a629e1539d87fa47afee8847a253690437-1614799468655.pdf"})
```

```python
#No works
test({"username":"tester"})

test({"username":"bleh", "token":token, "text":"Hello database! (again)","file":"tester-636120a629e1539d87fa47afee8847a253690437-1614799468655.pdf"})

test({"username":"tester", "text":"Hello database! (again)", "token":token})

test({"username":"tester", "token":token, "text":"Hello database! (again)","file":"tester-636120a629e1539d87fa47afee8847a253690437-1614799468654.pdf"})

test({"username":"tester", "file":"tester-636120a629e1539d87fa47afee8847a253690437-1614799468655.pdf"})
```
